```python
filename = "Downloads/stress_analysis.csv"
names = ['Mean HR','SDNN','RMSSD','NN50','PNN50','SD1','SD2','ApEn',
         'VLF(ms^2)','LF(ms^2)','HF(ms^2)','LF/HF ratio','TP',
         'VLF_lomb(ms^2)','LF_lomb(ms^2)','HF_lomb(ms^2)','LF/HF_lomb ratio','TP_lomb','Stress']
df = read_csv(filename, names = names)
array = df.values
X = np.array(df.iloc[:,0:18])
y = np.array(df.iloc[:,18])
y = y.reshape(-1,1)
scaler = MinMaxScaler(feature_range = (0,1))
rescaled_X = scaler.fit_transform(X)
rescaled_y = scaler.fit_transform(y)
```

```python
model = Sequential()
model.add(Dense(12, input_dim = 18, kernel_initializer = 'normal', activation = 'relu'))
model.add(Dense(8, kernel_initializer = 'normal', activation = 'relu'))
model.add(Dense(1, kernel_initializer = 'normal', activation = 'sigmoid'))

# model = Sequential()
# model.add(Dense(12, input_dim = 5, kernel_initializer = 'normal', activation='relu'))
# model.add(Dense(8, kernel_initializer = 'normal', activation = 'relu'))
# model.add(Dense(1, kernel_initializer = 'normal', activation = 'sigmoid' ))

```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(rescaled_X, rescaled_y, validation_split = 0.2, nb_epoch = 100, batch_size = 10)

scores = model.evaluate(rescaled_X, rescaled_y)

print("%s: %.2f%% " % (model.metrics_names[1], scores[1]*100))


# model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# model.fit(X, y, validation_split = 0.33, nb_epoch = 100, batch_size = 10)

# scores = model.evaluate(X, y)

# print("%s: %.2f%% " % (model.metrics_names[1], scores[1]*100))

# model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# # Fit the model
# model.fit(X, y,validation_split=0.33, nb_epoch=30, batch_size=10)
# # evaluate the model
# scores = model.evaluate(X, y)
# print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
X_train, X_test, y_train, y_test = train_test_split(rescaled_X, rescaled_y, test_size = 0.2, random_state = 7)
model = Sequential()
model.add(Dense(12, input_dim = 18, kernel_initializer = 'normal', activation = 'relu'))
model.add(Dense(8, kernel_initializer = 'normal', activation = 'relu'))
model.add(Dense(1, kernel_initializer = 'normal', activation = 'sigmoid'))

model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X_train, y_train, validation_data = (X_test, y_test), nb_epoch = 150, batch_size = 10)
scores = model.evaluate(X_train, y_train)
print("%s: %.2f%% " % (model.metrics_names[1], scores[1]*100))
```

```python
# K cross test validation
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)
cvscores = []
for train, test in kfold.split(rescaled_X, rescaled_y):

    model = Sequential()
    model.add(Dense(12, input_dim = 18, kernel_initializer='normal', activation = 'relu'))
    model.add(Dense(8, kernel_initializer='normal', activation='sigmoid'))
    model.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))
#compile model
    model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
#Fit the model
    model.fit(X_train, y_train, nb_epoch = 150, batch_size= 10, verbose=0)
#evaluate the model
    scores = model.evaluate(X_test, y_test, verbose=0)
    print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
    cvscores.append(scores[1] * 100)
print("%.2f%% (+/- %.2f%%)" % (np.mean(cvscores), np.std(cvscores)))
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
```

```python
# Function to create the model for Keras Classifier
def create_model():
    model = Sequential()
    model.add(Dense(12, input_dim=18, kernel_initializer='normal', activation='relu'))
    model.add(Dense(8, kernel_initializer='normal', activation = 'relu'))
    model.add(Dense(1, kernel_initializer='normal', activation = 'sigmoid'))
# Compile model

    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model

model = KerasClassifier(build_fn=create_model, nb_epoch=150, batch_size=10, verbose=0)
# evaluate using 10-fold cross validation
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=7)
results = cross_val_score(model, rescaled_X, rescaled_y, cv=kfold)
print(results.mean())
```
